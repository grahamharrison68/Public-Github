```python
import numpy as np
from matplotlib.pyplot import scatter

#xx = np.array([-0.51, 51.2])
#yy = np.array([0.33, 51.6])
xx = np.array([0, 100])
yy = np.array([0, 20])
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7         # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] 

m = np.random.multivariate_normal(means, covs, 1000).T
scatter(m[0], m[1], s=1, c="red")
```

```python
def make_data(x_min, x_max, y_min, y_max, corr, size, std = 3):
    xx = np.array([x_min, x_max])
    yy = np.array([y_min, y_max])
    means = [xx.mean(), yy.mean()]  

    stds = [xx.std() / std, yy.std() / std]

    covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
            [stds[0]*stds[1]*corr,           stds[1]**2]] 

    return np.random.multivariate_normal(means, covs, size).T
```

```python
import matplotlib.pyplot as plt
```

```python
d1 = make_data(6, 9, 6, 9, -0.9, 1000)
d2 = make_data(5, 11, 5, 11, -0.9, 1000)
d3 = make_data(5.5, 12, 5.5, 12, -0.9, 1000)
d4 = make_data(6, 13, 6, 13, -0.9, 1000)
d5 = make_data(8, 12, 8, 12, -0.9, 1000)
d6 = make_data(9, 12, 9, 12, -0.9, 1000)

scatter(d1[0], d1[1], s=1, c="red")
scatter(d2[0], d2[1], s=1, c="blue")
scatter(d3[0], d3[1], s=1, c="green")
scatter(d4[0], d4[1], s=1, c="cyan")
scatter(d5[0], d5[1], s=1, c="yellow")
scatter(d6[0], d6[1], s=1, c="purple")

plt.show()
```

```python
d_all = np.concatenate((d1, d2, d3, d4, d5, d6), axis=1)
```

```python
scatter(d_all[0], d_all[1], s=1, c="black")
```

```python
from scipy import stats

def draw(x, y, s=2, c="black"):
    slope, intercept, r, p, std_err = stats.linregress(x, y)

    def myfunc(x):
      return slope * x + intercept

    mymodel = list(map(myfunc, x))

    plt.plot(x, mymodel, c=c)
    return plt.scatter(x, y, s=1, c=c)
```

```python
plt.figure(figsize=(10, 7.5)) 
plt.xlabel("Exercise")
plt.ylabel("Cholestorol")

draw(d_all[0], d_all[1], 2, "black")
plt.show()
```

```python
plt.figure(figsize=(10, 7.5)) 
plt.xlabel("Exercise")
plt.ylabel("Cholestorol")

a = draw(d1[0], d1[1], s=1, c="red")
b = draw(d2[0], d2[1], s=1, c="blue")
c = draw(d3[0], d3[1], s=1, c="green")
d = draw(d4[0], d4[1], s=1, c="cyan")
e = draw(d5[0], d5[1], s=1, c="orange")
f = draw(d6[0], d6[1], s=1, c="purple")

plt.legend([a, b, c, d, e, f],['20-25', '26-30', '31-35', '36-40', '41-45', '46-50'], loc = 'upper left')

plt.show()
```
