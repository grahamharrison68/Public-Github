 # Categorical Values

 We can deduce that Family, Education, Personal Loan, Securities Account, CD Account, Online and CreditCard are categorical.
 However, Mortgage is almost certainly not categorical a sit has 347 values ...

```python
print(df_bank['Family'].value_counts())
print(df_bank['Education'].value_counts())
print(df_bank['Mortgage'].value_counts())
print(df_bank['Personal Loan'].value_counts())
print(df_bank['Securities Account'].value_counts())
print(df_bank['CD Account'].value_counts())
print(df_bank['Online'].value_counts())
print(df_bank['CreditCard'].value_counts())
print(df_bank['Mortgage'].value_counts())
```

```python
sns.countplot(df_bank['Personal Loan'])
plt.show()
sns.countplot(df_bank['Education'])
plt.show()
```
