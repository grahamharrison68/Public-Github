### Introduction
Before free text can be properly analysed or fed into a machine learning algorithm it will nead to be "cleaned". 

Data cleaning is standard stage of the data science pipeline and it involves various cleaning operations, for example replacing null values with a sensible non-null value, resolving outlying values that could skew the results etc. but when the input data is free text there are additional ways to clean it.

For example, white space can be removed, punctuation can be taken out, text can be changed to lower case etc. 

Another type of cleaning operation that can be applied to free text is expanding contractions. For example - "I've had a great day today, how's your day been?" contains the contractions "I've" and "how's" which can be expanded into "I have" and "how has".

The point of expanding the contractions is that further down the data science pipeline when the text is tokenized (converted into numeric values) so that it can be used in a machine learning algorithm the words "I've" will have a different tokenized values to the words "I have" and this will cause problems because these words have the same meaning.

The solution is to replace all of the contractions with their unabbreviated, expanded equivalents but when I looked into this online there was a problem.

It appears that there has previously been a library called ``contractions`` but it will not install with ``pip install contractions`` or the conda equivalent and a bit of research shows that this library is no longer supported.
