## Building the Directed Acyclic Graph (DAG)
The structure of a DAG can be inferred by using one of several programmatic causal discovery techniques or by utilising the expertise of domain experts. 

That process is beyond the scope of this article so for now we will assume that the structure of the DAG for greenhouse gases in Italy is as follows -

```python
from pgmpy.models import BayesianNetwork

climate_model = BayesianNetwork([("POP", "EC"),   ("URB", "EC"),   ("GDP", "EC"),
                                 ("EC", "FFEC"),  ("EC", "REC"),   ("EC", "EI"),
                                 ("REC", "CO2"),  ("REC", "CH4"),  ("REC", "N2O"),
                                 ("FFEC", "CO2"), ("FFEC", "CH4"), ("FFEC", "N2O")])
```

At this stage we have defined the DAG in by creating a ``BayesianNetwork`` that holds the links between the nodes but it is very difficult to visualise how the nodes are connected and what is causing what.

### "Daft" Network Diagrams
My first thought was to utilize the "daft" diagrams that are implemented directly within the ``BayesianNetwork`` class in the ``pgmpy`` library as they are quick and easy to use and the look-and-feel is very similar to the examples in "The Book of Why" by Judea Pearl & Dana Mackenzie -

```python
import matplotlib.pyplot as plt

drug_model = BayesianNetwork([("Blood Pressure", "Heart Attack"), ("Drug", "Blood Pressure"), ("Drug", "Heart Attack")])
drug_model.to_daft().render()

plt.show()
```

https://www.amazon.co.uk/Book-Why-Science-Cause-Effect/dp/0241242630

So far, so good, but look what happens when a more complex network like the Italian greenhouse gases example is rendered in daft ...

```python
climate_model.to_daft().render()
plt.show()
```

The default layout for daft charts is "circular". It is possible to change the default and also to change the node positions but however you modify the parameters the results are not inadequate for complex networks.

The following code snippets are written as calls into my ``dag_tools`` library. Full Access to the library together with all source code and documentation is given at the end of the article so you can easily create DAGs for any causal inference model yourself.

### NetworkX Diagrams

The drawbacks of daft network diagrams lead me to exploring another graphical library to render the DAG as a ``networkx`` chart (https://networkx.org/) -

```python
dag_tools.display_networkx_model(model=climate_model) 
```

The default layout for ``networkx`` diagrams is the spring layout which makes an attempt to layout the nodes that has not worked very well for this example as some nodes are drawn on top of each other making it impossible to read and interpret.

My next attempt involved extending my ``dag_tools`` library to explicitly specify the node positions ...

```python
POS : dict = {'POP': [0, 0], 'URB': [1, 0], 'GDP': [2, 0], "EC": [3, 0], "FFEC": [4, 0], "EI": [0, 1], "REC": [1, 1], "CO2" : [2, 1], "CH4" : [3, 1], "N2O": [4, 1]}

dag_tools.display_networkx_model(model=climate_model, pos=POS) 
```

Specifying the node positions means that the nodes do not overlay each other and hence we can really start to visualise and understand the causes and effects.

However, there is still room for improvement. 

In specifying the node positions explicitly we almost needed to understand how the nodes are connected before the visualisation can take place.

This drawback led to me extending the ``dag_tools`` library with a layout algorithm of my own so that the caller just has to specify how many columns they want the nodes arranging into ...

```python
dag_tools.display_networkx_model(model=climate_model, auto_layout_cols=4)
```

This is a better visulaisation of the DAG that did not require any pre-analysis of the raw data that has been created in a single line of Python code.

In this arrangement for example it is easy to see that FFEC (Fossil Fuel Consumption) causes CO2 (Carbon Di-oxide emissions), CH4 (Methan emissions) and N2O (Nitrous Oxide emissions).

There is one big drawback with this approach however, it is not "intractive".

### Pyvis Diagrams
I had seen various tutorials online relating to ``pyvis`` network diagrams which lead me to wonder if I could put a Python wrapper around ``pyvis`` to enable directed acyclic graphs to be explored and positioned interactively.

This led to the implementation of ``display_pyvis_model`` in my ``dag_tools`` library which took a long time to get right, particularly because I really wanted to be able to fully render the network diagrams in a cell within a Jupyter Notebook.

Here are the final results ...

```python
dag_tools.display_pyvis_model(model=climate_model, enable_physics=False, hierarchical=False, notebook=True)
```

This is a highly graphical, intuitive and easy-to-use way to explore the structure of a directed acyclic graph and to make sense 
of the relationships between the nodes.

It is very simple to untangle the relationships and to visualise and understnad the causes and effects by simply dragging the nodes around until the arrows are not crossing.

Howeverm there are still a few issues -

- ``pyvis`` is difficult to control and the documentation is complicated (using ``dag_tools`` will resolve this).
- The initial layout is inconsistent and uncontrollable; it will be different every time you run the code.
- It would be very nice to be able to save the final positions of the nodes but there is no way to do this.

However ``dag_tools.display_pyvis_model`` provides a great way to visualise DAGs that I have used successfully many times in production projects and that the customers have really liked.
